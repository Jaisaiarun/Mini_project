```python
import pandas as pd
```

```python
data = pd.read_csv("statlog.csv")#dataset before preprocessing
display(data.head())
```

```python
data.isnull().sum()#Null values identification
```

```python
#Dropping rows with Null value
print("Shape before droping: ")
print(data.shape)
data=data.dropna()
print("Shape after droping: ")
print(data.shape)
print(data.isnull().sum())
```

```python
#Changing the multiclass label to binary
data['num'] = data.num.map({1: 0, 2: 1})
display(data)
#df['thal'] = df.thal.fillna(df.thal.mean())
#df['ca'] = df.ca.fillna(df.ca.mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
num_count  = data['num'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(num_count.index, num_count.values, alpha=0.8)
plt.title('Heart Disease Count')
plt.ylabel('Count', fontsize=12)
plt.xlabel('Heart Disease ', fontsize=12)
plt.show()
```
